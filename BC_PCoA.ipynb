```python
import numpy as np
import pandas as pd
import scipy as sp


data =  pd.read_csv('~/Downloads/sushmita_data.csv',index_col=0, header=0, sep=',')
```

```python
data
```

```python
import scipy.spatial
data_2 = data.values
ids = data.index
bc_dm = sp.spatial.distance.squareform(sp.spatial.distance.pdist(data_2,"braycurtis"))
```

```python
print(bc_dm)
```

```python
from skbio.stats.ordination import pcoa
bc_pc = pcoa(np.nan_to_num(bc_dm))
```

```python
import matplotlib.pyplot as plt
fig = bc_pc.plot()
```
